```python
# Create a big decimal number
big_dec_num = 1000 ** 1000

# Convert it to a hex string
big_hex_num = hex(big_dec_num)

# Output the length of the hex string
print(f"Num of Hex Chars: {len(big_hex_num)}")

# Convert the hex string back to decimal and compare to the original decimal value
print(big_dec_num == int(big_hex_num, 16))

# Note: Python 2 has a maxint, Python 3 does not. So any size int regardless of base should work fine.

# print(big_dec_num)
# print(big_hex_num)
```
